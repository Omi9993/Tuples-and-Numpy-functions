```python
# Create two sets
set1 = {1, 2, 3, 4, 5}
set2 = {2, 3, 7, 6, 1}
```

```python
# Union of two sets.
Result = set1.union(set2)
```

```python
print(Result)
```

```python
# intersection of two sets.
Result1 = set1.intersection(set2)
```

```python
print(Result1)
```

```python
# difference of two tuples.
Result2 = set1.difference(set2)
```

```python
print(Result2)
```

Create numpy program to append values to the end of an array

```python
import numpy as np
Array1 = [10, 20, 30]
print("Original array:", Array1)

Array1 = np.append(Array1, [[40, 50, 60], [70, 80, 90]])
print("After append values to the end of the array:", Array1)
```
